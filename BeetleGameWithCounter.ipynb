```python
#Build Beetle Function With Counter
def buildbeetle():
    #Die Random Number Generator
    import random

    #Die Throw Counter
    throw_count = 0
    count_one = 0
    count_two = 0
    count_three = 0
    count_four = 0
    count_five = 0
    count_six = 0
    
    #Beetle Builder
    beetle = 'start'
    step = 0

    body = 0
    head = 0
    antenna = 0
    eye = 0
    mouth = 0
    leg = 0

    print('WELCOME TO THE BEETLE GAME\n')

    while beetle != 'complete':
        #Check if Beetle is completely built
        if (step > 0) and (antenna == 2) and (eye == 2) and (mouth == 1) and (leg == 6):
            beetle = 'complete'
            print('CONGRATULATIONS!!! BEETLE IS COMPLETELY BUILT...')
            print('SUMMARY OF DIE THROWN: \nYou threw your die', throw_count,'times.')
            print('You threw 1', count_one,'times.')
            print('You threw 2', count_two,'times.')
            print('You threw 3', count_three,'times.')
            print('You threw 4', count_four,'times.')
            print('You threw 5', count_five,'times.')
            print('You threw 6', count_six,'times.')
            break
        
        #Keep Throwing Die
        print('Throw Your Die')
        die = random.randint(1,6)
        throw_count += 1
        print('Player throws',die)

        #Die Checker and Corresponding Move
        if (die == 1):
            count_one += 1
            if body < 1 :
                body += 1
                print('Collect Body!')
                step = 1
                print('step',step,'... Body is complete!\n')
            else: print('Body is Already Available! Go Again!\n')
        elif (die == 2) and (step > 0):
            count_two += 1
            if head < 1 :
                head += 1
                print('Collect Head!')
                step = 2
                print('step',step,'... Head is complete!\n')
            else: print('Head is Already Available! Go Again!\n')
        elif (die == 3) and (step == 2):
            count_three += 1
            if antenna < 2:
                antenna += 1
                print('Collect Antenna',antenna,'!\n')
            else: print('Antenna is complete!\n')
        elif (die == 4) and (step == 2):
            count_four += 1
            if eye < 2:
                eye += 1
                print('Collect Eye',eye,'!\n')
            else: print('Eyes are complete!\n')
        elif (die == 5) and (step == 2):
            count_five += 1
            if mouth < 1 :
                mouth += 1
                print('Collect Mouth!\n')
            else: print('Mouth is Already Available! Go Again!\n')
        elif (die == 6) and (step > 0):
            count_six += 1
            if leg < 6 :
                leg += 1
                print('Collect Leg!.. You now have',leg,'leg(s)\n')
            else: print('Legs are Complete! Go Again!\n')
        else:
            if(die==1): count_one += 1 
            if(die==2): count_two += 1
            if(die==3): count_three += 1
            if(die==4): count_four += 1
            if(die==5): count_five += 1
            if(die==6): count_six += 1
            print("You can't take anything. Go Again!\n")
            continue
        
            
#Call Beetle Function
buildbeetle()
```
